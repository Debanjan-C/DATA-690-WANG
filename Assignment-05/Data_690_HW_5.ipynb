### 9. Find out the definition of the above 9 variables and document the definitions in the Jupyter Notebook in a Markdown cell so that the readers of your notebook will know the meaning of these variables.

#### DEFINITIONS
- **UNITID** - This is the unit id for institutions reported in the academic year of 2017 - 2018 *(September to June)* for integrated post-secondary data collection year of 2017 - 2018.
- **INSTNM** - This is the institution namea for the academic year 2017 - 2018 that are reported in integrated post-secondary data collection year of 2017 - 2018, unless more of the recent datas are avilable. Currently, the most recent data reported is for the Academic year of 2019 - 2020
- **CITY** - This is the list of cities mentioned in the file for the academic year of 2017 - 2018 reported in the integrated post-secondary data collection year - 2017 to 2018. More recent datas are reported and for the year of 2019 - 2020.
- **STABBR** - This is the list of states mentioned in the file for the academic year of 2017 - 2018 reported in the integrated post-secondary data collection year - 2017 to 2018.
- **ZIP** = This is the data that contains zip codes of the institutions mentioned int he file for the academic year of 2017 - 2018 that is reported in the integrated post-secondary data collection for year 2017-2018.
- **ADM_RATE** - The administrative rate for the fall of 2017  reported in the integrated post-secondary data collection for year 2017-2018.
- **UGDS** =  The Enrollment of Undergrad for Fall 2017. Reported in the integrated post - secondary  data collection for the year of 2017-2018
= **TUITIONFEE_IN** - In state Tuition fee for  the academic year of 2017- 2018 reported in the integrated post - secondary  data collection for the year of 2017-2018.
- **MN_EARN_WNE_P10** -  The mean or the average earnings of students working and not enrolled 10 years after entry
